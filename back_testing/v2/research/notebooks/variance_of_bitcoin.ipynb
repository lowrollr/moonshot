<h3>Variance of Bitcoin</h3>
<p> Here we are calculating the variance of bitcoin to see if it can help give any insight to how a coin might move. By studying the variance of a coin it can also give some insight on when to enter or not enter the market. </p>


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_df = pd.read_csv('../../historical_data/XRPUSD_1.csv')
my_df.columns = ['time', 'open', 'high', 'low', 'close', 'volume', 'trades']
# my_df = my_df[(my_df.time > 1512086400) & (my_df.time < 1512432000)]
my_df['rolling_variance'] = my_df['close'].rolling(120).var()
```

```python
prices = my_df['close']
variance = my_df['rolling_variance']

fig, ax1 = plt.subplots()

ax1.plot(my_df['time'], prices, color='red')
ax2 = ax1.twinx()

ax2.plot(my_df['time'], variance, color='blue')
fig.tight_layout()
plt.show()

```
